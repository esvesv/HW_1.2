```python
import pandas as pd
```

```python
df = pd.read_csv('data/test_data.csv')
```

```python
df
```

```python
thresh = df.shape[0] * 0.5

df.dropna(axis=1, thresh=df.shape[0] - thresh, inplace=True)

df.dropna(axis=0, thresh=df.shape[1] - 2, inplace=True)

values = {
            'one': df.one.mean(),
            'two': df.two.mean(),
            'tree': df.three.mean(),
            'four': df.four.mode()[0]
}

df = df.fillna(values)
```
